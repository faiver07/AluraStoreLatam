### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda.info()
```

#1. Análisis de facturación



```python
ingresos_total_tienda1 = tienda['Precio'].sum()
print(f'Cantidad de ingresos de tienda 1: $ {ingresos_total_tienda1}')
ingresos_total_tienda2 = tienda2['Precio'].sum()
print(f'Cantidad de ingresos de tienda 2: $ {ingresos_total_tienda2}')
ingresos_total_tienda3 = tienda3['Precio'].sum()
print(f'Cantidad de ingresos de tienda 3: $ {ingresos_total_tienda3}')
ingresos_total_tienda4 = tienda4['Precio'].sum()
print(f'Cantidad de ingresos de tienda 4: $ {ingresos_total_tienda4}')
```

```python
import matplotlib.pyplot as plt

# Calcular ingresos totales por tienda
ingresos = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}

# Graficar
plt.figure(figsize=(8,5))
plt.bar(ingresos.keys(), ingresos.values(), color='skyblue')
plt.title('Ingresos totales por tienda')
plt.ylabel('Ingresos ($)')
plt.xlabel('Tienda')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
def productos_por_categoria(tienda, nombre_tienda):
    conteo = tienda['Categoría del Producto'].value_counts().reset_index()
    conteo.columns = ['Categoría', 'Cantidad de Productos']

    print(f'\n{nombre_tienda}: Categorías más vendidas')
    print(conteo)
    return conteo

# Aplicar la función a cada tienda
productos_por_categoria(tienda, 'Tienda 1')
productos_por_categoria(tienda2, 'Tienda 2')
productos_por_categoria(tienda3, 'Tienda 3')
productos_por_categoria(tienda4, 'Tienda 4')


```

```python
# Conteo de productos por categoría
categoria_t1 = tienda['Categoría del Producto'].value_counts()

# Graficar
plt.figure(figsize=(6,6))
plt.pie(categoria_t1, labels=categoria_t1.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de categorías - Tienda 1')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio(tienda, nombre_tienda):
    promedio = tienda['Calificación'].mean()
    print(f'Calificación promedio en {nombre_tienda}: {round(promedio, 2)}')
    return promedio

# Aplicar la función a cada tienda
calificacion_promedio(tienda, 'Tienda 1')
calificacion_promedio(tienda2, 'Tienda 2')
calificacion_promedio(tienda3, 'Tienda 3')
calificacion_promedio(tienda4, 'Tienda 4')
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

def analizar_ventas_productos(tienda, nombre_tienda):
    conteo = tienda['Producto'].value_counts()

    # Mostrar el producto más y menos vendido
    mas_vendido = conteo.idxmax()
    menos_vendido = conteo.idxmin()
    print(f'\n{nombre_tienda}:')
    print(f'Producto más vendido: {mas_vendido} ({conteo.max()} ventas)')
    print(f'Producto menos vendido: {menos_vendido} ({conteo.min()} venta(s))')

    # Visualizar los 10 productos más vendidos
    top10 = conteo.head(10)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=top10.values, y=top10.index, palette="viridis")
    plt.title(f'{nombre_tienda} - Top 10 productos más vendidos')
    plt.xlabel('Cantidad vendida')
    plt.ylabel('Producto')
    plt.tight_layout()
    plt.show()

# Aplicar a cada tienda
analizar_ventas_productos(tienda, 'Tienda 1')
analizar_ventas_productos(tienda2, 'Tienda 2')
analizar_ventas_productos(tienda3, 'Tienda 3')
analizar_ventas_productos(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
def costo_envio_promedio(tienda, nombre_tienda):
    promedio_envio = tienda['Costo de envío'].mean()
    print(f'Costo de envío promedio en {nombre_tienda}: ${round(promedio_envio, 2)}')
    return promedio_envio

# Aplicar a cada tienda
costo_envio_promedio(tienda, 'Tienda 1')
costo_envio_promedio(tienda2, 'Tienda 2')
costo_envio_promedio(tienda3, 'Tienda 3')
costo_envio_promedio(tienda4, 'Tienda 4')
```

# **EXTRA:Análisis del desempeño geográfico**

```python
import matplotlib.pyplot as plt

# Crear un dataframe combinado con una columna de tienda
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

ventas_geo = pd.concat([tienda, tienda2, tienda3, tienda4])

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
for tienda_name, df in ventas_geo.groupby('Tienda'):
    plt.scatter(df['lon'], df['lat'], label=tienda_name, alpha=0.5, s=15)

plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Distribución geográfica de ventas por tienda')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
